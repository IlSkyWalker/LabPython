### Розв’язати систему за допомогою формул Крамера і виконати перевірку за допомогою:
*  **матричного множення;**
    * Нехай для матриці $А$ порядку $n на n$ існує зворотна матриця $A^{-1}$. Помножимо обидві частини матричного рівняння $A ⋅ X =В$ зліва на $A^{-1}$ (порядки матриць $A ⋅ X =В$ дозволяють зробити таку операцію, дивіться статтю операції над матрицями, властивості операцій). Маємо $A^{-1}(A ⋅ X) =A^{-1}⋅ B$. Так як для операції множення матриць відповідних порядків характерно властивість асоціативності, то останню рівність можна переписати як $(A^{-1}⋅A) ⋅ X =A^{-1}⋅ B$, а за визначенням оберненої матриці $A^{-1}⋅A = E$ ($E$- одинична матриця порядку $n на n$), тому
$$(A^{-1}⋅A) ⋅ X =A^{-1}⋅ B$$
$$E⋅X = A^{-1}⋅A $$
$$X =A^{-1}⋅B$$
      Таким чином, рішення системи лінійних алгебраїчних рівнянь матричним методом визначається за формулою $X =A^{-1}⋅B$ . Іншими словами, рішення СЛАР знаходиться за допомогою оберненої матриці $A^{-1}$.

### Рішення з допомогою формул Крамера
Метод Крамера полягає в тому, що ми послідовно знаходимо головний визначник системи (5.1), тобто визначник матриці $А$

$$D = det (ai j)$$

   і $n$ допоміжних визначників $D$ i $(i = 1, ..., n)$, які виходять з визначника $D$ заміною $i-го$ стовпця стовпцем вільних членів.

   Формули Крамера мають вигляд:

$$D × x i = D i (i = 1, ..., n). (5.2)$$

   З (5.2) випливає правило Крамера, яке дає вичерпну відповідь на питання про спільності системи (5.1): якщо головний визначник системи відмінний від нуля, то система має єдине рішення, яке визначається за формулами:

$$x i = D i / D.$$

   Якщо головний визначник системи $D$ і всі допоміжні визначники $D i = 0 (i = 1, ..., n)$, то система має безліч рішень. Якщо головний визначник системи $D = 0$, а хоча б один допоміжний визначник відмінний від нуля, то система несумісна.
### Обчислити значення матричного виразу.
Дано матричні рівняння:
        $$AA·X = B, (1)$$
        $$Y·A = B, (2)$$
де $A$ і $B$ - задані матриці, причому $A$ - невироджена. Потрібно знайти       матриці $X$ і $Y$.
Згадаємо, що в числовій області частка від ділення b на a визначається як рішення рівняння $a ∙ x = b$ $(або x ∙ a = b)$ і існує не завжди. Можна спробувати визначити «поділ» матриць, розглядаючи рівняння (1) і (2), в яких, згідно з правилом множення, матриці $A, B, X, Y$ не можуть мати довільну структуру. Так, в першому рівнянні матриці A і B повинні мати однакове число рядків, а в другому - однакове число стовпців. Вже звідси ясно, що якщо навіть обидва ці рівняння однозначно розв'язні (а це далеко не завжди так), то їх рішення цілком можуть бути матрицями не тільки різними, але і різної структури. Таким чином, для матриць виявляється неможливим визначити розподіл зі звичними властивостями. Матричні рівняння виду (1) і (2) вирішуються наступним чином. Так як A невироджена матриця, то існує обернена матриця A-1. Помножимо зліва обидві частини рівняння (1) на $A^{-1}$: $A^{-1} · A · X = A^{-1} · B$, тоді отримаємо $E · X = A^{-1} $· B, або $$X = A^{-1}$$ · B. (3) Аналогічно, множачи справа обидві частини рівності (2) на$ A^{-1}$, будемо мати: $Y · A · A^{-1} = B · A^{-1}$, звідки знаходимо, що $$Y = B · A^{-1}. (4)$$

```python
print("Крамер")
import numpy
A= numpy.array([[2,-1, 1, -1], [2, -1, 0, -3], [3, 0, -1, 1],[2, 2, -2, 5]]) # Матрица (левая часть системы)
B = numpy.array([1, 2, -3,-6]) # Вектор (правая часть системы)
print("Матричне множення")
Aob = numpy.linalg.inv(A)
print("Обратная матрица \n",Aob,"\n")
print("X= B*A^(-1) \n",(Aob.dot(B)),"\n")
print(" numpy.linalg.solve().")
numpy.linalg.solve(A, B)
np.allclose((Aob.dot(B)), np.linalg.solve(A, B), rtol=0, atol=1e-7)
```

```python
import numpy as np
print("Розв’язати рівняння ")
A = np.array([[5, -1, 3], [0, 2, -1], [-2, -1, 0]])
B = np.array([[3, 7, -2], [1, 1, -2], [0, 1, 3]]) 
print("Матричне множення")
print("A= \n",A,"\n")
print("B= \n",B,"\n")
detA = np.linalg.det(A)
print("Det",detA)
a11 = A[1,1]*A[2,2]-A[2,1]*A[1,2]
a12 = -1*(A[1,0]*A[2,2]-A[2,0]*A[1,2])
a13 = A[1,0]*A[2,1]-A[2,0]*A[1,1]
a21 = -1*(A[0,1]*A[2,2]-A[2,1]*A[0,2])
a22 = A[0,0]*A[2,2]-A[2,0]*A[0,2]
a23 = -1*(A[0,0]*A[2,1]-A[2,0]*A[0,1])
a31 = A[0,1]*A[1,2]-A[1,1]*A[0,2]
a32 = -1*(A[0,0]*A[1,2]-A[1,0]*A[0,2])
a33 = A[0,0]*A[1,1]-A[1,0]*A[0,1]
Ad = np.array([[a11,a12,a13],[a21,a22,a23],[a31,a32,a33]])
print("Матрица алгебраических дополнений\n",Ad)
At = Ad.transpose()
print("Транспонированная матрица\n",At)
Aob1 = (1/5)*At
print("Обратная матрица(Вручную)(At*(1/5))\n",Aob1,"\n")
Aob2 = numpy.linalg.inv(A)
print("Обратная матрица (numpy.linalg.inv())\n",Aob2,"\n")
print("(Вручную) X= B*A^(-1) \n",Aob1.dot(B))
print("X= B*A^(-1) \n",Aob2.dot(B))
print("\n numpy.linalg.solve().")
print("X= \n",np.linalg.solve(A, B))
np.allclose(Aob2.dot(B), np.linalg.solve(A, B), rtol=0, atol=1e-7)
```
