```python
 def task_2(num1, num2):
    while num1 != 0 and num2 != 0:
        if num1 >= num2:
            num1 -= num2
        else:
            num2 -= num1
    return num1 or num2
def is_number(str):
    try:
        float(str)
        return True
    except ValueError:
        return False
print(" ")
try:
    a,b = input().split()
    if is_number(a) and is_number(b) == True:
        answer=task_2(int(a),int(b))
        print("    {0}  {1} - ".format(a,b),answer)
    else:
        print("   ")
except:
    print("Error")



```
